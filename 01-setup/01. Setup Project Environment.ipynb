## Set-up the project environment for Formula1 Project
1. Create External Location

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS databricks_course_ext_dl1_formula1_incr
URL 'abfss://formula1-incr@dbrickscoursestgacct.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `databricks-course-sc`)
COMMENT 'External location for the formula1-incr container';
```

```sql
%fs
ls 'abfss://formula1-incr@dbrickscoursestgacct.dfs.core.windows.net/'
```

### Create Catalogs

```sql
SHOW CATALOGS;
```

```sql
CREATE CATALOG IF NOT EXISTS formula1_incr
MANAGED LOCATION 'abfss://formula1-incr@dbrickscoursestgacct.dfs.core.windows.net/'
COMMENT 'This is the main catalog for the formula1-incr project';
```

### Create Schemas

```sql
CREATE SCHEMA IF NOT EXISTS formula1_incr.landing;
CREATE SCHEMA IF NOT EXISTS formula1_incr.bronze
    MANAGED LOCATION 'abfss://formula1-incr@dbrickscoursestgacct.dfs.core.windows.net/bronze';
CREATE SCHEMA IF NOT EXISTS formula1_incr.silver
    MANAGED LOCATION 'abfss://formula1-incr@dbrickscoursestgacct.dfs.core.windows.net/silver';
CREATE SCHEMA IF NOT EXISTS formula1_incr.gold
    MANAGED LOCATION 'abfss://formula1-incr@dbrickscoursestgacct.dfs.core.windows.net/gold';
```

```sql
USE CATALOG formula1_incr;
SELECT current_catalog();
```

```sql
SHOW SCHEMAS;
```

### Create Volume Files

```sql
CREATE EXTERNAL VOLUME formula1_incr.landing.files
LOCATION 'abfss://formula1-incr@dbrickscoursestgacct.dfs.core.windows.net/landing';
```

```sql
%fs
ls /Volumes/formula1_incr/landing/files
```
